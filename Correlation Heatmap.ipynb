```python
import pandas as pd
```

```python
df = pd.read_csv("../data/data100.csv")
len(df.columns)
```

```python
df.drop("hard_timeout", axis=1, inplace=True)
df.drop("flags", axis=1, inplace=True)
df.drop("idle_timeout", axis=1, inplace=True)
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool_))
correlation_matrix = df.corr()

heatmap = sns.heatmap(correlation_matrix, mask=mask, vmin=-1, vmax=1, annot=True, fmt=".2f", cmap='coolwarm')

# Add the correlation values to the squares of the lower triangle
for i in range(correlation_matrix.shape[0]):
    for j in range(correlation_matrix.shape[1]):
        if i > j:
            text = f"{correlation_matrix.iloc[i, j]:.2f}"
            heatmap.annotate(text, (j + 0.5, i + 0.5), ha='center', va='center', color='black')

heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize': 18}, pad=16)

plt.show()
```
