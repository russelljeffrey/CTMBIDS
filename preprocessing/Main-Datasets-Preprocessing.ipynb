# data18_sub_features_binary_classification

```python
df = pd.read_csv("../data/data18.csv")
df18 = df.copy()
```

```python
# We first define categorical and numerical features
cat_features = ["datapath_id", "ip_src", "ip_dst", "ip_proto", "icmp_code", "icmp_type", "flags"] # categorical features
num_features = ["flow_id", "tp_src", "tp_dst", "flow_duration_sec", "flow_duration_nsec", "idle_timeout", "hard_timeout", "packet_count", "byte_count", "packet_count_per_second", "packet_count_per_nsecond", "byte_count_per_second", "byte_count_per_nsecond"] #numerical features
number_of_features = len(cat_features) + len(num_features)
```

```python
selected_features = ['datapath_id', # selected feature are specified in Recursive Feature Elimination Notebook
 'flow_id',
 'tp_src',
 'tp_dst',
 'ip_proto',
 'icmp_code',
 'icmp_type',
 'flow_duration_sec',
 'flow_duration_nsec',
 'idle_timeout',
 'packet_count',
 'byte_count',
 'packet_count_per_second',
 'packet_count_per_nsecond',
 'byte_count_per_second',
 'byte_count_per_nsecond']
```

```python
# Min-Max Function
def min_max_normalize(column):
    min_val = column.min()
    max_val = column.max()
    column_normalized = (column - min_val) / (max_val - min_val)
    return column_normalized
```

```python
x = pd.DataFrame() # a new empty dataframe. We will add preprocessed data into x dataframe. 

# We move forward Column by column to see which column is categorical or numerical.
for i in range(number_of_features):
    feature = df18.columns[i]
    if feature in cat_features and feature in selected_features:
        x = pd.concat([x, pd.get_dummies(df18[feature], prefix=feature, dtype='int8')],axis=1)
    elif feature in num_features and feature in selected_features:
        # Apply min-max normalization to the specified column
        x = pd.concat([x, min_max_normalize(df18[feature])],axis=1)
x
```

```python
x = pd.concat([x, df18["label"]], axis=1)
x.fillna(0, inplace=True)
x.to_csv("../data/preprocessed/data18_sub_features_binary_classification.csv", index=False)
```
